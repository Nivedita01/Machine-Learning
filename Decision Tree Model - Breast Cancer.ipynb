# Decision Tree Classifier
## What are Decision Trees?

Decision trees are widely used models for classification and regression tasks. Essentially, they learn a hierarchy of if/else questions, leading to a decision. Decision trees are assigned to the information-based learning algorithms which use different measures of information gain for learning. 

A decision tree mainly contains of a root node, interior nodes, and leaf nodes which are then connected by branches.

#### Importing the required Libraries

```python
# Importing the required Libraries 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.metrics import confusion_matrix 
from sklearn.cross_validation import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
```

#### Importing our Cancer Dataset from Scikit Learn Databases

```python
# Importing the Breast Cancer Dataset
from sklearn.datasets import load_breast_cancer
cancer_dataset = load_breast_cancer()
```

##### Examining the Data (Keys, Description, Type, Shape, Head)

```python
# Examining the Data
# Printing the Keys of Iris Dataset
print("Keys of iris_dataset: \n{}".format(cancer_dataset.keys()))

# Printing the Descriptions of Iris Dataset
print(cancer_dataset['DESCR'][:193] + "\n...")

# Type of Data
print("Type of data: {}".format(type(cancer_dataset['data'])))

# Quick check of how many instances (rows) and how many attributes (columns) the data contains with the shape property.
#iris_dataset.shape
print("Shape of data: {}".format(cancer_dataset['data'].shape))

# Head gives the first/top few records of the Dataset
#iris_dataset.head(5)
print("First five rows of data:\n{}".format(cancer_dataset['data'][:5]))
```

##### Examining the Target (Type, Shape, Names, Features)

```python
# Model Prediction Target - Species of the flower that we want to predict
print("Target names: {}".format(cancer_dataset['target_names']))

# Type of the Target
print("Type of target: {}".format(type(cancer_dataset['target'])))

# Shape of the Target 
print("Shape of target: {}".format(cancer_dataset['target'].shape))

# Features - Descriptions of each Feature
print("Feature names: \n{}".format(cancer_dataset['feature_names']))
```

#### Create Training and Testing Data (using Scikit-Learn)

```python
# Importing Scikit-Learn function to create Training and Testing sets
# Shuffle the Data and Split into Training and Testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
cancer_dataset['data'], cancer_dataset['target'], random_state=42)
```

#### Examining Training and Testing Data

```python
# Shape of Training Data
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))

# Shape of Testing Data
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

#### Buidling the Decision Tree Classification Model

```python
# Building the Decision Tree Classification Machine Learning Algorithm
# Fitting clasifier to the Training set
# Loading libraries
from sklearn.metrics import confusion_matrix 
from sklearn.cross_validation import train_test_split 
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report

# Instantiate learning model
tree = DecisionTreeClassifier(random_state =0)

# Fitting the model
tree.fit(X_train, y_train)

# Printing the Accuracy of Training and Testing Sets
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

The accuracy on the training set is 100%—because the leaves are pure, the tree was grown deep enough that it could perfectly memorize all the labels on the training data.

Since we didnot restrict the depth of a decision tree, the tree can be deep and complex.Unpruned trees are prone to overfitting and not generalizing well to new data.

We can apply pre-pruning to the tree, which will stop developing the tree before we perfectly fit to the training data. Hence, limiting the tree to a certain depth is going to stop developing the tree once it reaches that certain depth.

Limiting the depth of the tree decreases overfitting. This leads to a lower accuracy on the training set, but an improvement on the test set.

#### Applying Prepruning techniques and Evaluating Accuracy Scores

```python
# Applying Pre Pruning to our Descision Tree
tree = DecisionTreeClassifier(max_depth=4, random_state=0)
tree.fit(X_train, y_train)

# Printing the Accuracy of Training and Testing Sets
print("Accuracy on training set: {:.3f}".format(tree.score(X_train, y_train)))
print("Accuracy on test set: {:.3f}".format(tree.score(X_test, y_test)))
```

#### Analyzing Feature Importance in Decision Trees

```python
# Feature Importance in Decision Trees
# It defines how important each feature is for the decision a tree makes
# Number between 0 and 1 for each feature where 0 is not used at all and 1 means "perfectly predicts the Target"

print("Feature importances:\n{}".format(tree.feature_importances_))

# Visualizing the Features

def plot_feature_importances_cancer(model):
    n_features = cancer_dataset.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer_dataset.feature_names)
    plt.xlabel("Feature importance")
    plt.ylabel("Feature")
    plt.ylim(-1, n_features)
plot_feature_importances_cancer(tree)
```

#### Similarities and Differences between DT and KNN

1.	Decision trees can be used to predict both continuous and discrete values i.e. they work well for both regression and classification tasks.KNN is used for clustering and classification.

2. KNN determines neighborhoods, so there must be a distance metric. This implies that all features must be numeric. Distance metrics may be affected by varying scales between attributes and high-dimensional space.DT on the other hand predicts a class for a given input vector. The attributes may be numeric or nominal.

3. So, if we want to find similar examples we can use KNN. However, If we want to classify examples we could use DT.

4.	Since KNN performs on-the-spot learning, it requires frequent database lookups, hence, can be computationally expensive. Decision Tree Classifier does not require such lookups as it has in-memory classification model ready.

5.	Since KNN performs instance-based learning, a well-tuned K can model complex decision spaces having arbitrarily complicated decision boundaries, which are not easily modeled by other "eager" learners like Decision Trees.

6. Classifiers like Decision Tree, Bayesian, Back-propagation, Support Vector Machine come under the category of "Eager Learners", because they first build a classification model on the training dataset before being able to actually classify an unseen observation from test dataset. The learned model is now "eager" (read hungry) to classify previously unseen observations.

7. The KNN-based classifier, however, does not build any classification model. It directly learns from the training instances (observations). It starts processing data only after it is given a test observation to classify. Thus, KNN comes under the category of "Lazy Learner" approaches.

8.	"Eager" learners work in batches, modeling one group of training observations at a time. So they are not fit for incremental learning. But KNN naturally supports incremental learning (data streams) since it is an instance-based learner.

9.	Decision Trees are very fast and efficient compared to KNN and other classification algorithms.

10. Decision Trees can be ensembled (combined into multiple ML models) to create very powerful models like Randomforests, GradientBoost and XGBoost.
